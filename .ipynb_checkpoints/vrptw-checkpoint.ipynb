```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from itertools import islice
```

```python
file = 'aliahmadi-et-al-2021-set-1/p1.txt'
with open(file, 'r') as text:
    textfile = text.read()
    print(textfile)
```

```python
text_row = np.array(textfile.split('\n')
)
```

```python
# node
df_nodes = pd.DataFrame(
    data = [text_row[i].split('\t') for i in range(2)],
    columns = ['n_debots', 'n_bins', 'n_interm_facilites']
)
```

```python
df_nodes
```

```python
def convert(lst, var_lst):
    it = iter(lst)
    return [list(islice(it, i)) for i in var_lst]
```

```python
def splitting(data):
    res = []
    for ele in data.split(' '):
        if ele.strip():
            res.append(ele)
    return res
def preprocess(data, start_row, end_row, shape):
    temp = [data[i].split('\n') for i in range(start_row, end_row)]
    result = []
    for i in range(end_row - start_row):
        result = np.append(result, splitting(temp[i][0]))
    return np.array(list(convert(result, shape)), dtype=object)
```

```python
# activify
df_activity1 = pd.DataFrame(preprocess(text_row, 3, 10, [7 for x in range(7)]),
                   columns = ['n_trucks', 'n_trailers1', 'n_trailers2', 'n_trailers3', 'n_trailers3', 'n_twpd1', 'n_twpd2'])
df_activity1
```

```python
df_activity2 = pd.DataFrame(preprocess(text_row, 10, 19, [7 for x in range(7)]),
                   columns = ['n_trucks', 'n_trailers1', 'n_trailers2', 'n_trailers3', 'n_trailers3', 'n_twpd1', 'n_twpd2'])
df_activity2
```

```python
# route node
df_route1 = pd.DataFrame(preprocess(text_row, 31, 39, [2 for x in range(7)]),
                   columns = ['i', 'j'])
df_route1
```

```python
# LN: The number of traffic light between nodes i,j .

df_tlamp = pd.DataFrame(preprocess(text_row, 39, 46, [2 for x in range(7)]),
                       columns = ['n_tlamp1', 'm_tlamp2'])
df_tlamp
```

```python
# ET: Exiting time of the vehicle v.

df_time = pd.DataFrame(preprocess(text_row, 47, 54, [2 for x in range(7)]),
                      columns = ['time1', 'time2'])
df_time
```

```python
# R:  An uncertain amount of MSW in node i and TW s.

df_umsw1 = pd.DataFrame(preprocess(text_row, 56, 58, [2 for x in range(2)]),
                      columns = ['umsw1', 'umsw2'])
df_umsw1
```

```python
df_umsw2 = pd.DataFrame(preprocess(text_row, 59, 61, [2 for x in range(1)]),
                      columns = ['umsw1', 'umsw2'])
df_umsw2
```
